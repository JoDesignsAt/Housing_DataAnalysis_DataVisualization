# Housing Data Model Training

```python
df.head()
```

```python
X = df.drop(['price','date'], axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)  # no fitting for test data because not giving prior info
```

# Creating a Model using Tensorflow Keras DeepLearning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x = X_train, y = y_train, 
          validation_data = (X_test, y_test), batch_size=128, epochs=400)
```

# Evaluating a Model

```python
pd.DataFrame(model.history.history)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

### No increase in validation_loss over train_loss, no over fitting, so continue training..

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)  # average absolute error across predictions
```

```python
mean_squared_error(y_test, predictions) # average error across predictions, 
                                        # equals val_loss: 27098168011.5358
```

```python
np.sqrt(mean_squared_error(y_test, predictions)) 
```

```python
explained_variance_score(y_test, predictions)
```
